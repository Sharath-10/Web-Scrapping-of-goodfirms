```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

company_name = []
no_reviews=[]
net_charge=[]
founded_in=[]
employees_capacity=[]
located_in=[]
final_price=[]
for i in range(1,5):
    url = "https://www.goodfirms.co/directory/languages/top-software-development-companies?page="+str(i)
    PATH = "C:\chromedriver.exe"
    driver = webdriver.Chrome(PATH)
    driver.get(url)
    driver.maximize_window()
    companies = driver.find_elements(By.XPATH,'//div[@class="service-entity"]')
    for company in companies: 
        names = driver.find_elements(By.XPATH,'//div[@class="listing-name-tag"]')
        for name in names:
            company_name.append(name.text)
        charges=driver.find_elements(By.XPATH,'//div[@class="firm-pricing"]')
        for charge in charges:
            net_charge.append(charge.text)
        found=company.find_elements(By.XPATH,'//div[@class="firm-founded"]')
        for year in found:
            founded_in.append(year.text)
        capacity=company.find_elements(By.XPATH,'//div[@class="firm-employees"]')
        for cap in capacity:
            employees_capacity.append(cap.text)
        locate=company.find_elements(By.XPATH,'//div[@class="firm-location"]')
        for loc in locate:
            located_in.append(loc.text)
            
        

        try:
            if len(company.find_elements(By.XPATH,".//span[@class='listinv_review_label']"))>0:
                reviews = company.find_elements(By.XPATH,".//span[@class='listinv_review_label']")
                for review in reviews:
                    # print('the length is===>', len(review.text), review.text)
                    no_reviews.append(review.text)
            else:
                no_reviews.append("0")
        except:
            pass

    
print('no of companies==>',len(company_name))
print('no of reviews==>',len(no_reviews))
print(net_charge)
print(founded_in)
print(employees_capacity)
print(located_in)
import pandas as pd
df = pd.DataFrame(zip(company_name,no_reviews,net_charge,founded_in,employees_capacity,located_in),columns=['company_name','no_reviews','net_charge','founded_in','employees_capacity','located_in'])

df.to_excel(r"d:\ass.xlsx",index=False)

driver.quit()
```
